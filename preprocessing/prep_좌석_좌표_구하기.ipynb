```python
# 맨 처음 부채꼴 원의 중심 설정
center_x, center_y = 0, 0

# 부채꼴 정보 설정
start_angle = 254.5
angle_span = 31
num_points = 52
radius = 5

# 데이터를 저장할 리스트 초기화
all_points = []

# 부채꼴 생성
angle_step = angle_span / (num_points - 1)
points = []

for j in range(num_points):
    angle = np.deg2rad(start_angle + j * angle_step)  # 각도를 라디안으로 변환
    x = center_x + radius * np.cos(angle)
    y = center_y + radius * np.sin(angle)
    points.append((x, y))

all_points.extend(points)

# 2-11행 추가 부채꼴 정보 설정
additional_radii = [5.054, 5.108, 5.162, 5.216, 5.27, 5.324, 5.378, 5.432, 5.486, 5.54] #추가 반지름
additional_num_points = [54, 54, 56, 56, 59, 59, 61, 61, 61, 64] # 찍어야 하는 점 개수

for i in range(len(additional_radii)):
    radius = additional_radii[i]
    num_points = additional_num_points[i]

    angle_step = angle_span / (num_points - 1)
    points = []

    for j in range(num_points):
        angle = np.deg2rad(start_angle + j * angle_step)
        x = center_x + radius * np.cos(angle)
        y = center_y + radius * np.sin(angle)
        points.append((x, y))

    all_points.extend(points)

# 12-22행 추가 부채꼴 정보 설정(13,14행 사이 복도 포함)
additional_start_angle = [254, 254, 254, 254, 254, 254.5, 254.5, 255, 255.5, 256, 256.5, 256.5]
additional_angle_span = [32, 32, 32, 32, 32, 31, 31, 30, 29, 28, 27, 27]
additional_radii = [5.594, 5.648, 5.702, 5.756, 5.81, 5.864, 5.918, 5.972, 6.026, 6.08, 6.134, 6.188]
additional_num_points = [66, 66, 66, 66, 68, 67, 67, 65, 65, 63, 62, 62]

for i in range(len(additional_radii)):
    radius = additional_radii[i]
    num_points = additional_num_points[i]

    additional_angle_step = additional_angle_span[i] / (num_points - 1)
    points = []

    for j in range(num_points):
        angle = np.deg2rad(additional_start_angle[i] + j * additional_angle_step)
        x = center_x + radius * np.cos(angle)
        y = center_y + radius * np.sin(angle)
        points.append((x, y))

    all_points.extend(points)

# 데이터프레임 생성
data = {'x': [point[0] for point in all_points],
        'y': [point[1] for point in all_points]}
F1_ABCDE = pd.DataFrame(data)

# CSV 파일로 저장
F1_ABCDE.to_csv('point_ABCDE.csv', index=True)

F1_ABCDE.head()
```
