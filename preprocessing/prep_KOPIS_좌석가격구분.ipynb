```python
import pandas as pd

seat_df = pd.read_csv("/content/drive/MyDrive/2023빅콘/data/예술의전당/kopis_concert.csv")
seat_df.head(3)
```

```python
seat_df['pcseguidance'].unique()
```

```python
import re
import numpy as np

# 가격만 추출하는 함수
def extract_prices(text):
    if pd.isna(text):
        return []
    return [int(re.sub(",", "", price[:-1])) for price in re.findall(r'(\d+,\d+원)', text)]

seat_df['prices'] = seat_df['pcseguidance'].apply(extract_prices)

# R S A 가격 배정
def get_high(prices):
  return max(prices) if prices else np.nan

def get_low(prices):
  return min(prices) if prices else np.nan

def get_mid(prices):
  return (max(prices) + min(prices)) // 2 if prices else np.nan

seat_df['R'] = seat_df['prices'].apply(get_high)
seat_df['S'] = seat_df['prices'].apply(get_mid)
seat_df['A'] = seat_df['prices'].apply(get_low)
```
