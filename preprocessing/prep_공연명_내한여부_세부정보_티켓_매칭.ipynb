```python
import pandas as pd

# 결과를 저장할 빈 데이터프레임 생성
expanded_concert_price = pd.DataFrame()

# 각 행에 대해 'prfpdfrom'과 'prfpdto' 사이의 날짜 범위를 구함
for idx, row in concert_price.iterrows():
    date_range = pd.date_range(row['prfpdfrom'], row['prfpdto'])
    for date in date_range:
        new_row = row.copy()
        new_row['play_date'] = date.strftime('%Y%m%d') # 날짜 형식을 '20220206'과 같이 변환
        expanded_concert_price = expanded_concert_price.append(new_row, ignore_index=True)

# 'play_date'를 문자열 타입으로 변환 (필요한 경우)
expanded_concert_price['play_date'] = expanded_concert_price['play_date'].astype(str)
```

```python
memb_val = merged_df[['membership_type_1', 'membership_type_2', 'membership_type_3',
                        'membership_type_4', 'membership_type_5', 'membership_type_6']].values.tolist()

merged_df['membership'] = memb_val
merged_df['membership'] = merged_df['membership'].apply(lambda x: [val for val in x if not pd.isna(val)])

def custom_sort(lst):
    order = ['무료', '싹틔우미', '노블', '골드', '블루', '그린']
    return sorted(lst, key=lambda x: order.index(x) if x in order else len(order))

merged_df['membership'] = merged_df['membership'].apply(custom_sort)

unique_memberships = ['무료', '싹틔우미', '노블', '골드', '블루', '그린']

for membership_type in unique_memberships:
    merged_df[membership_type] = merged_df['membership'].apply(lambda x: 1 if membership_type in x else 0)
```

```python
import pandas as pd

# 주어진 조건에 따라 값을 추출하는 함수
def parse_seat_info(seat_str):
    parts = seat_str.split()

    # floor 처리
    if '합창석' in parts[0]:
        floor = 0
    else:
        floor = int(parts[0].split('층')[0])

    # block 처리
    if 'BOX' in parts[1]:
        block = int(parts[1].split('BOX')[1])
        row = 0
    elif '블록' in parts[1]:
        block = parts[1].split('블록')[0]
        row = int(parts[1].split('열')[0].split('블록')[1])
    else:
        block = None
        row = None

    # number 처리
    number = int(parts[2])

    return pd.Series([floor, block, row, number], index=['floor', 'block', 'row', 'number'])
```

```python
# apply 함수를 사용하여 새로운 컬럼 생성
sub_df = merged_df['좌석'].apply(parse_seat_info)
```

```python
sub_df.rename(columns={'floor': '층', 'block': '블록', 'row': '열', 'number': '좌석번호'}, inplace=True)
```
