```python
# 공연명을 저장할 빈 리스트 생성
matching_performance_names = []

# data_concert의 각 행에 대해 공연명 찾기
for index, row in data_concert.iterrows():
    play_date = row['play_date']
    matching_names = kopis_concert[(kopis_concert['시작일'] <= play_date) & (play_date <= kopis_concert['종료일'])]['공연명'].tolist()
    
    if matching_names:
        matching_performance_names.append(', '.join(matching_names))
    else:
        matching_performance_names.append(np.nan)

# 결과를 data_concert 데이터프레임에 추가
data_concert['공연명'] = matching_performance_names

# 결과 출력
data_concert
```
