# 공연 목록 및 상세 데이터 긁기

## 1. 공연목록 조회 서비스

```python
ServiceKey = 'fbc17f20fa8d4a8d8a130b025019ae3a'
url = f'http://www.kopis.or.kr/openApi/restful/pblprfr?service={ServiceKey}&stdate=20180101&eddate=20230731&rows=1000000&cpage=1&prfplccd=FC000001-03'
```

```python
response = requests.get(url)
contents = response.text

pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
# XML을 OrderedDict로 변환
data_dict = xmltodict.parse(contents)
```

```python
print(data_dict.keys())
print(data_dict['dbs'].keys())
print(data_dict['dbs']['db'][0])
print(len(data_dict['dbs']['db']))
```

```python
## fcltynm이 '예술의전당'인 것만 mt20id 긁기
fcltynm_list = []

for data in data_dict['dbs']['db']:
  if data['fcltynm'] == '예술의전당':
    fcltynm_list.append(data['mt20id'])
```

## 2. 공연 상세 조회 서비스

```python
ServiceKey = 'fbc17f20fa8d4a8d8a130b025019ae3a'
prfrID = 'PF219388' # ex. PF132236
url = f'http://www.kopis.or.kr/openApi/restful/pblprfr/{prfrID}?service={ServiceKey}'
```

```python
response = requests.get(url)
contents = response.text

pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
# XML을 OrderedDict로 변환
data_dict = xmltodict.parse(contents)
```

```python
ServiceKey = 'fbc17f20fa8d4a8d8a130b025019ae3a'
```

```python
## fcltynm이 '예술의전당 (콘서트홀)'인 것만 데이터 긁기
concert_db = []

for id in tqdm(fcltynm_list):
  prfrID = id
  url = f'http://www.kopis.or.kr/openApi/restful/pblprfr/{prfrID}?service={ServiceKey}'

  response = requests.get(url)
  contents = response.text

  data_dict = xmltodict.parse(contents)

  if data_dict['dbs']['db']['fcltynm'] == '예술의전당 (콘서트홀)':
    prf_info = {}
    prf_info['prfnm'] = data_dict['dbs']['db']['prfnm'] # 공연명
    prf_info['prfpdfrom'] = data_dict['dbs']['db']['prfpdfrom'] # 시작일
    prf_info['prfpdto'] = data_dict['dbs']['db']['prfpdto'] # 종료일
    prf_info['prfcast'] = data_dict['dbs']['db']['prfcast'] # 공연출연진
    prf_info['prfcrew'] = data_dict['dbs']['db']['prfcrew'] # 공연제작진
    prf_info['prfruntime'] = data_dict['dbs']['db']['prfruntime'] # 공연 런타임
    prf_info['prfage'] = data_dict['dbs']['db']['prfage'] # 공연 관람 연령
    prf_info['entrpsnm'] = data_dict['dbs']['db']['entrpsnm'] # 공연 제작진
    prf_info['pcseguidance'] = data_dict['dbs']['db']['pcseguidance'] # 티켓가격
    prf_info['sty'] = data_dict['dbs']['db']['sty'] # 줄거리 -> 필요한가?
    prf_info['genrenm'] = data_dict['dbs']['db']['genrenm'] # 공연장르명
    prf_info['prfstate'] = data_dict['dbs']['db']['prfstate'] # 공연상태 -> 필요한가?
    prf_info['dtguidance'] = data_dict['dbs']['db']['dtguidance'] # 공연시간

    concert_db.append(prf_info)
```
