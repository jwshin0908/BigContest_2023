```python
plt.figure(figsize=(15,5))
sns.countplot(x='age', data=data, order=data['age'].unique().tolist().sort(key=lambda x: str(x)[:2]))
plt.xlabel("나이대")
plt.ylabel("예매 건수")
plt.title("나이대별 예매정보")
plt.show()
```

### running_time : 러닝 타임(단위 : 분)

```python
data_temp = data.copy()

# 조건에 따라 값을 변경
data_temp['running_time'] = data_temp['running_time'].apply(lambda x: f'{(x // 30) * 30}~{(x // 30) * 30 + 30}' if x < 360 else x)

data_temp
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='running_time', data=data_temp, order=data_temp['running_time'].unique().tolist())
plt.xlabel("러닝 타임")
plt.ylabel("건수")
plt.title("러닝 타임별 건수")
plt.show()
```
